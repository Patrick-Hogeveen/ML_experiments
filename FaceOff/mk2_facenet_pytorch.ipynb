```python
from facenet_pytorch import MTCNN, InceptionResnetV1
import cv2
import numpy as np
import torch
import argparse
import utils
from PIL import Image
import matplotlib.pyplot as plt

# If required, create a face detection pipeline using MTCNN:
mtcnn = MTCNN(keep_all=True)

# Create an inception resnet (in eval mode):
resnet = InceptionResnetV1(pretrained='vggface2').eval()

device = torch.device('cuda' if torch.cuda.is_available else 'cpu')
print(device)
```

```python
img = Image.open("data/1*8WXOiaNfAAYFVySOgMTRhg.jpeg").convert('RGB')

img_array = np.array(img, dtype=np.float32)
img_array = cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)

bounding_boxes, conf, landmarks = mtcnn.detect(img, landmarks=True)
```

```python
# draw the bounding boxes around the faces
image_array = utils.draw_bbox(bounding_boxes, img_array)
# plot the facial landmarks
image_array = utils.plot_landmarks(landmarks, img_array)
```

```python
image_rgb = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb/255.0)
```
